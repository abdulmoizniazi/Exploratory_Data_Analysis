## 8. Detecting Outliers

```
#Detect outlier for cylinders
```

```
numcol = df.select_dtypes(include='number').columns
iqr_series = pd.Series(dtype='float64')
for i, col in enumerate(numcol):
    IQR = df[col].quantile(0.75) - df[col].quantile(0.25)
    iqr_series[col] = IQR.round(1)

print(f"{iqr_series}")
# for k, v in iqr_series.items():
#     print(f"{k}\t \t {v:.2f}")
```

```
#print IQR after finding IQR
```

Don't worry about the above values because it's not important to know each and every one of them because it's just important to know how to use this technique in order to remove the outliers.

```
numeric_cols = df.select_dtypes(include=['number']).columns

# Create a DataFrame to store Q1, Q3, and IQR values
stats_df = pd.DataFrame(index=numeric_cols, columns=['Q1', 'Q3', 'IQR'])

# Calculate Q1, Q3, and IQR for each numeric column
for column in numeric_cols:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    # Store the values in the stats_df DataFrame
    stats_df.loc[column, 'Q1'] = Q1
    stats_df.loc[column, 'Q3'] = Q3
    stats_df.loc[column, 'IQR'] = IQR

# Print the statistics DataFrame
print(stats_df)

# Filter out outliers for each numeric column
for column in numeric_cols:
    Q1 = stats_df.loc[column, 'Q1']
    Q3 = stats_df.loc[column, 'Q3']
    IQR = stats_df.loc[column, 'IQR']

    # Filter out outliers
    df = df[~((df[column] < (Q1 - 1.5 * IQR)) | (df[column] > (Q3 + 1.5 * IQR)))]

# Print the shape of the filtered DataFrame
print(f"Filtered DataFrame shape: {df.shape}")

# Create boxplots for the filtered DataFrame
# plt.figure(figsize=(15, 10))
for i, column in enumerate(numeric_cols):
    plt.subplot(len(numeric_cols), 1, i + 1)
    sns.boxplot(x=df[column])
    plt.title(f'Boxplot of {column} after outlier removal')

plt.tight_layout()
plt.show()
```

## 9. Plot different features against one another (scatter), against frequency (histogram)

### Histogram

Histogram refers to the frequency of occurrence of variables in an interval. In this case, there are mainly 10 different types of car manufacturing companies, but it is often important to know who has the most number of cars. To do this histogram is one of the trivial solutions which lets us know the total number of car manufactured by a different company.

```
for i, column in enumerate(numeric_cols):
    plt.subplot(len(numeric_cols), 1, i + 1)
    sns.histplot(df[column], bins=20, kde=True)  # KDE line for smoothness
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
```

### Heat Maps

Heat Maps is a type of plot which is necessary when we need to find the dependent variables. One of the best way to find the relationship between the features can be done using heat maps. In the below heat map we know that the price feature depends mainly on the Engine Size, Horsepower, and Cylinders.

```
numeric_df = df.select_dtypes(include=['number'])
correlation_matrix = numeric_df.corr()
print(correlation_matrix)
# Plot the heatmap
# plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```
