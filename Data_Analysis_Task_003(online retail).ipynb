```python
numcol = df.select_dtypes(include=[np.number]).columns
for col in numcol:
    mean_val = df[col].mean()
    df[col].fillna(mean_val, inplace= True) # df[col] = df[col].fillna(mean_val)
df.isna().sum()
```

```python
strcol = df.select_dtypes(exclude=np.number).columns
for col in strcol:
    mode = df[col].mode()[0]
    df[col] = df[col].fillna(mode)
df.isna().sum()
```

```python
plt.hist(df['Country'], bins=20, edgecolor='black')
plt.title('Histogram with 20 Bins between 20 and 100')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(df['UnitPrice'], bins=20, kde=False, edgecolor='black')

# Add title and labels
plt.title('Distribution of Unit Price', fontsize=18)
plt.xlabel('Different Unit Price for different items', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Show plot
plt.show()
```

```python
df_top20 = df.sort_values(by='Quantity', ascending=False).head(20)
print(df_top20)
# Plot the data

sns.barplot(x=df_top20['Country'], y=df_top20['Quantity'], data=df_top20, palette='viridis')

# Add labels and title
plt.title('Top 20 Countries in Online Retail Sales', fontsize=16)
plt.xlabel('Quantity', fontsize=14)
plt.ylabel('Country', fontsize=14)

# Show plot
plt.tight_layout()
plt.show()
```
