```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import pandas as pd
import time
```

```python
def web_driver():
    options = webdriver.ChromeOptions()
    options.add_argument("--verbose")
    options.add_argument('--no-sandbox')
    options.add_argument('--headless')
    options.add_argument('--disable-gpu')
    options.add_argument("--window-size=1920, 1200")
    options.add_argument('--disable-dev-shm-usage')
    driver = webdriver.Chrome(options=options)
    return driver
```

```python
base_url = "https://aitoptools.com/ai-tools/"
# category = ['business-productivity', 'tech-development', 'design-media', 'lifestyle-growth', 'writing-content', 'customer-service', 'chatbot', 'image-generator', 'education', 'writing-assistant']
category = ['business-productivity', 'tech-development']

for c in category:
  names_list = []
  descrip_list = []
  tool_link_list = []
  monthly_visits_list = []
  bounce_rate_list = []
  visit_duration_list = []
  pages_per_visit_list = []
  country_rank_list = []
  global_rank_list = []

  def main():
    driver = web_driver()
    driver.get(base_url + c)
    print(f"\n\n--------------------  {c} ------------------------\n\n")
    i = 1
    try:
        while i!=3:
            try:
                load_more_button = driver.find_element(By.ID, 'loadMore')
                ActionChains(driver).move_to_element(load_more_button).click(load_more_button).perform()
                time.sleep(5)
            except Exception as e:
                print("No Load More btn")
                break
            i += 1

        get_names = driver.find_elements(By.XPATH, "//div[@id='ajax-content']/div[@class='col-xs-12 col-sm-6 col-md-4 loop-block']/div[1]/div[2]/div[1]/h2")
        get_descrip = driver.find_elements(By.XPATH, "//div[@class='custom-listing-content']/p")
        get_tool_link = driver.find_elements(By.XPATH, "//div[@class='custom-listing']/a")

        for name in get_names:
            names_list.append(name.text)
        for description in get_descrip:
            descrip_list.append(description.text)
        for link in get_tool_link:
            tool_link_list.append(link.get_attribute('href'))

        driver.quit()

    except Exception as e:
        print("An error occurred:", e)


  XPATHS_For_Visits = [
      "//body/div[2]/section[2]/div[1]/div[1]/div[1]/div[5]/div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/strong",
      "//body/div[2]/section[2]/div[1]/div[1]/div[1]/div[7]/div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/strong"
  ]
  XPATHS_For_BounceRate = [
      "//body/div[2]/section[2]/div[1]/div[1]/div[1]/div[5]/div[1]/div[1]/div[1]/div[1]/div[2]/div[1]/div[1]/div[1]/div[1]/strong",
      "//body/div[2]/section[2]/div[1]/div[1]/div[1]/div[7]/div[1]/div[1]/div[1]/div[1]/div[2]/div[1]/div[1]/div[1]/div[1]/strong"
  ]
  XPATHS_For_VisitDuration = [
      "//body/div[2]/section[2]/div[1]/div[1]/div[1]/div[5]/div[1]/div[1]/div[1]/div[1]/div[3]/div[1]/div[1]/div[1]/div[1]/strong",
      "//body/div[2]/section[2]/div[1]/div[1]/div[1]/div[7]/div[1]/div[1]/div[1]/div[1]/div[3]/div[1]/div[1]/div[1]/div[1]/strong"
  ]
  XPATHS_For_PagesPerVisit = [
      "//body/div[2]/section[2]/div[1]/div[1]/div[1]/div[5]/div[1]/div[1]/div[1]/div[1]/div[4]/div[1]/div[1]/div[1]/div[1]/strong",
      "//body/div[2]/section[2]/div[1]/div[1]/div[1]/div[7]/div[1]/div[1]/div[1]/div[1]/div[4]/div[1]/div[1]/div[1]/div[1]/strong"
  ]
  XPATHS_For_CountryRank = [
      "//body/div[2]/section[2]/div[1]/div[1]/div[1]/div[5]/div[1]/div[1]/div[1]/div[1]/div[5]/div[1]/div[1]/div[1]/div[1]/strong",
      "//body/div[2]/section[2]/div[1]/div[1]/div[1]/div[7]/div[1]/div[1]/div[1]/div[1]/div[5]/div[1]/div[1]/div[1]/div[1]/strong"
  ]
  XPATHS_For_GlobalRank = [
      "//body/div[2]/section[2]/div[1]/div[1]/div[1]/div[5]/div[1]/div[1]/div[1]/div[1]/div[6]/div[1]/div[1]/div[1]/div[1]/strong",
      "//body/div[2]/section[2]/div[1]/div[1]/div[1]/div[7]/div[1]/div[1]/div[1]/div[1]/div[6]/div[1]/div[1]/div[1]/div[1]/strong"
  ]

  def main2():
      driver = web_driver()
      for link in tool_link_list:
          driver.get(link)
          visit_found = False
          rate_found = False
          duration_found = False
          pagepervisit_found= False
          countryRank_found= False
          globalRank_found= False
          #scraping monthly visits----
          for xpath in XPATHS_For_Visits:
              try:
                  visits = WebDriverWait(driver, 2).until(
                      EC.presence_of_element_located((By.XPATH, xpath))
                  )
                  monthly_visits_list.append(visits.text)
                  visit_found = True
                  break
              except TimeoutException:
                  continue
          if not visit_found:
              # print(f"Monthly visits not found for {link}")
              monthly_visits_list.append(" ")

          #scraping Bounce Rates----
          for xPath in XPATHS_For_BounceRate:
              try:
                  rates = WebDriverWait(driver, 2).until(
                      EC.presence_of_element_located((By.XPATH, xPath))
                  )
                  bounce_rate_list.append(rates.text)
                  rate_found = True
                  break
              except TimeoutException:
                  continue
          if not rate_found:
              # print(f"Bounce Rate not found for {link}")
              bounce_rate_list.append(" ")

          #scraping Visit Duration----
          for xPath in XPATHS_For_VisitDuration:
              try:
                  duration = WebDriverWait(driver, 2).until(
                      EC.presence_of_element_located((By.XPATH, xPath))
                  )
                  visit_duration_list.append(duration.text)
                  duration_found = True
                  break
              except TimeoutException:
                  continue
          if not duration_found:
              # print(f"Visit Duration not found for {link}")
              visit_duration_list.append(" ")

          #scraping PagesPerVisit----
          for xPath in XPATHS_For_PagesPerVisit:
              try:
                  page_per_visit = WebDriverWait(driver, 2).until(
                      EC.presence_of_element_located((By.XPATH, xPath))
                  )
                  pages_per_visit_list.append(page_per_visit.text)
                  pagepervisit_found = True
                  break
              except TimeoutException:
                  continue
          if not pagepervisit_found:
              # print(f"Page Per Visit not found for {link}")
              pages_per_visit_list.append(" ")

          #scraping Country Rank----
          for xPath in XPATHS_For_CountryRank:
              try:
                  rank = WebDriverWait(driver, 2).until(
                      EC.presence_of_element_located((By.XPATH, xPath))
                  )
                  country_rank_list.append(rank.text)
                  countryRank_found = True
                  break
              except TimeoutException:
                  continue
          if not countryRank_found:
              # print(f"CountryRank not found for {link}")
              country_rank_list.append(" ")

          #scraping Global Rank----
          for xPath in XPATHS_For_GlobalRank:
              try:
                  rank = WebDriverWait(driver, 2).until(
                      EC.presence_of_element_located((By.XPATH, xPath))
                  )
                  global_rank_list.append(rank.text)
                  globalRank_found = True
                  break
              except TimeoutException:
                  continue
          if not globalRank_found:
              # print(f"GlobalRank not found for {link}")
              global_rank_list.append(" ")



  main()
  # print("Reached to MAIN 2----------------------")
  main2()
  dic = {
        'Tool Name': names_list, 'Description': descrip_list,
        'Tool Link': tool_link_list,
        'Monthly Visits': monthly_visits_list,
        'Bounce Rate': bounce_rate_list,
        'Visit Duration': visit_duration_list,
        'Page Per Visit': visit_duration_list,
        'Country Rank': country_rank_list,
        'Global Rank': global_rank_list
  }
  df = pd.DataFrame(dic)
  df.to_excel(c + '.xlsx', index=False)
  print(df)
```
